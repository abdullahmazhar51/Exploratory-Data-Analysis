```python
## Function to find missing values percent in the dataset

def missingValues(df):
    missingcontent = round(df.isnull().sum()/len(df)*100,2)
    print("Total Missing Values Percentage in Dataframe:", round(missingcontent.mean(),2))
    print(missingcontent[missingcontent>0].sort_values(ascending=False))
```

```python
missingValues(df)
```

```python
df_country = pd.read_excel('countrycode.xlsx')
df_country
```

```python
##combining df and country df using join in pandas

final_df = pd.merge(df,df_country, on='Country Code', how = 'left')
```

```python
final_df.head(2)
```

```python
###Checking datatypes of df

final_df.dtypes
```

```python
final_df.columns
```

```python
## from this we can understand that max use of zomato is done in India
country_name = final_df.Country.value_counts().index
country_name
```

```python
country_value = final_df.Country.value_counts().values
country_value
```

```python
##Plotting country using zomato mostly.(Top 5 Country based on transaction)
plt.rcParams['figure.figsize'] = (16,8)
plt.pie(country_value[:4],labels = country_name[:4],autopct='%1.1f%%')
```

## Observertion 
1. Zomato maximum records or transaction are from India after that USA and then United Kingdom

```python
final_df.columns
```

```python
## Changing the name of 0th column

ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
ratings.head()
```

## Observation
1. When Rating is between 4.5 to 4.9 ---> Excellent
2. When Ratings are between 4.0 to 4.4 ---> Very Good
3. When Ratings are between 3.5 to 3.9 ---> Good
4. When Ratings are between 3.0 to 3.4 ---> Average
5. When Ratings are between 2.5 to 2.9 ---> Poor
6. When Ratings are between 2.0 to 2.4 ---> Very Poor

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = "Aggregate rating", y = "Rating Count",hue = 'Rating color', data= ratings, palette = ['white','red','orange','yellow','green','green'])
```

The above Graph looks like a Gaussian Curve which is good thing for any dataset

## Observation
1. Not rated value is very high
2. Max num of rating is between 2.5 to 3.9

```python
sns.countplot(data = ratings, x='Rating color', palette = ['white','red','orange','yellow','green','green'])
```

```python
## Finding the maximum very poor rating by country
final_df.groupby(['Aggregate rating','Country']).size().reset_index().rename(columns={0:'Very Poor'}).head(5)
```

## Observation

1. Maximum Very poor(0) rating given by india
